### 2. In which order will the expressions be evaluated.

`6 * 3 + 7 * 4`


<details>
  <summary>6*3 = 18. 7*4 = 28. 18 + 28 = 46. The multiplications takes precedence</summary>

  `*` > `*` > `+` 
</details>


`5 - 2 * 3 ** 4`

<details>
  <summary>3 ** 4 = 81 *2 = 162. Then 5 - 162 = -157</summary>

  `**` > `*` > `-`
</details>

`(5 - 2) * 3 ** 4`

<details>
  <summary>3**4 = 81 * (5-2) = 248</summary>

  `(-)` > `**` > `*` 
</details>

`5 + 2 >= 3 * 4`
